```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

### **Method 1**

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(
    penalty='l2',
    max_iter=500,
    eta0=0.1,
    learning_rate='constant',
    alpha=0.001
)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
print(f"r2 score = {r2_score(y_test, y_pred)}\n")
print(f"coef = {reg.coef_}\n")
print(f"intercept = {reg.intercept_}\n")
```

### **Method 2**

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
print(f"r2 score = {r2_score(y_test, y_pred)}\n")
print(f"coef = {reg.coef_}\n")
print(f"intercept = {reg.intercept_}\n")
```

### **From scratch**

```python
class ridgeGD:

    def __init__(self, epochs, learning_rate, alpha):
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.alpha = alpha
        self.coef = None
        self.intercept = None

    def fit(self, xtrain, ytrain):
        self.coef = np.ones(xtrain.shape[1])
        self.intercept = 0
        W = np.insert(self.coef, 0, self.intercept)
        xtrain = np.insert(xtrain,0,1,axis=1)
        for i in range(self.epochs):
            delLdelW = np.dot(xtrain.T, xtrain).dot(W) - np.dot(xtrain.T, ytrain) + self.alpha * W
            W = W - self.learning_rate * delLdelW
        self.coef = W[1:]
        self.intercept = W[0]

    def predict(self, xtest):
        return np.dot(xtest, self.coef) + self.intercept
```

```python
reg = ridgeGD(epochs=500, alpha=0.001, learning_rate=0.005)
```

```python
reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef)
print(reg.intercept)
```
