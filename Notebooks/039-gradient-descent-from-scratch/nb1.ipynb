### **Gradient Descent Step By Step**

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y = make_regression(
    n_samples = 4,
    n_features = 1,
    n_informative = 1,
    n_targets = 1,
    noise = 80,
    random_state = 13
)
```

```python
import matplotlib.pyplot as plot
plot.scatter(x,y)
```

### **Apply OLS**

To know value of slope and intercept in advance

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_   # slope
```

```python
reg.intercept_      # intercept
```

```python
plot.scatter(x,y)
plot.plot(x, reg.predict(x), color="#ff007b")
```

### **Applying Gradient Descent**

Assuming,

    m = 78.35

    b = 0

```python
y_pred = 78.35 * x + 0
```

```python
y_pred.reshape(4)
```

```python
plot.scatter(x,y)
plot.plot(x, reg.predict(x), color='#ff007b', label='OLS')
plot.plot(x, y_pred, color="#01c137", label='b = 0')
plot.legend()
plot.show()
```

### **Epoch 1**

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
lr = 0.1    # learning rate

step_size = lr * loss_slope
step_size
```

```python
# calculating new intercept

b = b - step_size
b
```

```python
y_pred1 = m * x + b

plot.scatter(x,y)
plot.plot(x, reg.predict(x), color='#ff007b', label='OLS')
plot.plot(x, y_pred, color="#01c137", label='b = 0')
plot.plot(x, y_pred1.ravel(), color="#d15bff", label='b = 20.92')
plot.legend()
plot.show()
```

### **Epoch 2**

```python
# Iteration 2

loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = m * x + b

plot.scatter(x,y)
plot.plot(x, reg.predict(x), color='#ff007b', label='OLS')
plot.plot(x, y_pred, color="#01c137", label='b = 0')
plot.plot(x, y_pred1.ravel(), color="#d15bff", label='b = 20.92')
plot.plot(x, y_pred2.ravel(), color="#5b66ff", label='b = 25.11')
plot.legend()
plot.show()
```

### **Epoch 3**

```python
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = m * x + b

plot.scatter(x,y)
plot.plot(x, reg.predict(x), color='#ff007b', label='OLS')
plot.plot(x, y_pred, color="#01c137", label='b = 0')
plot.plot(x, y_pred1.ravel(), color="#d15bff", label='b = 20.92')
plot.plot(x, y_pred2.ravel(), color="#5b66ff", label='b = 25.11')
plot.plot(x, y_pred3.ravel(), color="#ff975b", label='b = 25.95')
plot.legend()
plot.show()
```

From above plot, blue line with b = 25.11 is very close to our real best fit line with b = 25.15

### **Using loop**

```python
b = -100        # random value
m = 78.35       # we know this
lr = 0.01       # learning rate

epochs = 30

for i in range(epochs):
    slope = -2 * np.sum(y - m * x.ravel() - b)
    step_size = lr * slope
    b = b - step_size

    y_pred = m * x + b
    plot.plot(x, y_pred.ravel())

plot.scatter(x,y)
```
